# 필요한 모듈 import하기

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

# 데이터 준비

```python
cancer = load_breast_cancer()
cancer # data, target, frame, feature_names, target_names, images
```

```python
import pandas as pd

cancer_df = pd.DataFrame(data= cancer.data) #, row = digits.feature_name)
cancer_df
```

```python
cancer_data = cancer.data / 5000.0
print(cancer_data)
```

```python
cancer_label = cancer.target
print(cancer_label)
```

# train, test 데이터 분리

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_data, 
                                                    cancer_label, 
                                                    test_size=0.2, 
                                                    random_state=8)

print('X_train 개수: ', len(X_train), ', X_test 개수: ', len(X_test))
```

```python
X_train
```

```python
X_test
```

# 다양한 모델로 학습시켜보기

5가지의 모델로 학습을 시켜본 결과 거의 대부분이 좋은 성능을 보였지만, Random Forest 모델이 0.97의 정확도로 가장 높은 정확도를 보여줬다.


- Decision Tree

```python
# Decision Tree 이용
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
```

```python
y_pred = decision_tree.predict(X_test)
y_pred
```

```python
print(classification_report(y_test, y_pred))
# Decision Tree를 이용 : 0.94
```

- Random Forest

```python
# Random Forest 사용 : 0.97
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

- Support Vector Machine (SVM)

```python
# Support Vector Machine(SVM) 사용 : 0.94
from sklearn import svm
svm_model = svm.SVC()

svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

- Stochastic Gradient Descent Classifier(SGD Classifier)

```python
# Stochastic Gradient Descent Classifier(SGD Classifier) 사용 : 0.93
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

- Logistic Regression

```python
# Logistic Regression 사용 : 0.89
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter = 10000)

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

# 모델을 평가해보기

classification_report에는 precision, recall, f1-score 처럼 단순히 label 값을 예측한 accuracy보다 TP, TN, FP, FN을 활용하여 정확도를 보여준다.    
        
이중에서도 모델의 성능을 평가하는 지표로는 F1-score가 가장 좋다고 생각하는데, f1-score의 경우 2∗(precision∗recall)/precision+recall 이 공식을 바탕으로 구해지는 값이므로 f1-score가 높으면 precision과 recall도 자연스럽게 높은 수치를 기록하게 된다.

Precision은 내가 x라고 예측했을 때, 실제 데이터의 label이 x인 경우이고,    
Recall은 x라는 전체 label 갯수 중에서 내가 x라고 맞춘 비율이므로    
이 둘의 조화평균을 통해 구한 f1-score가 모델의 성능을 평가하는 지표로 사용되는 것이 좋다고 생각한다.
